```r
price = sample(seq(20,70,1.5),100,replace=TRUE)
```

```r
price
```

```r
data_gen = function(x) { return(0.35*x+rnorm(1)); }
c = 0 
area = NA
while ( c<= 100 ){
    area[c] = data_gen(price[c]); 
    c = c + 1 ; 
    }
```

```r
price
```

```r
area
```

```r
library(ggplot2)
```

```r
df = data.frame(price=price,area=area)
```

```r
ggplot(df,aes(x=area,y=price))+geom_point()+geom_smooth(aes(x=area,y=price))
```

```r
data = lm(df,x=area,y=price)
```
