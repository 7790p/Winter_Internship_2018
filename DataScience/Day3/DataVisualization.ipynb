```r
arm <- function(num) { 
		sum = 0 ; 
		temp = num;
		power = 0; 
		while (temp) { 
					   power = power + 1 ; 
					   temp = as.integer(temp/10);
					 }
	    #print(power)
		temp = num ;
		while (temp) { rem = temp %% 10; 
					   #print(rem);
					   sum = sum + rem^power;
                       #print(sum);					   
					   temp = as.integer(temp/10);
					   #print(temp);
				  }
		if ( num == sum ) { return(TRUE); }
		return (FALSE); 
	}
print(arm(1654))
```
