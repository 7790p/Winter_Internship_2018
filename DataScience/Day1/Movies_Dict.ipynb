```python
f = open('input/movies.dat')
d = f.readline()
f.seek(0)
print(d)
```

```python
types = []
for line in f : 
    d = line.split('|')
    d[0] = ''.join(d[0]).split("::")[-1]
    d[-1] = d[-1][:-1]
    types.append(d)

    
```

```python
types
```

```python
final_type =  []
for each_type in types : 
    final_type.extend(each_type)
print(final_type)
```

```python
types = list(set(final_type))
```

```python
types
```

```python
print("Total movie Types : ",len(types))
```

```python
movies = {}
for gener in types : 
    movies.setdefault(gener,dict())
```

```python
movies 
```

```python
years = []
f = open('input/movies.dat')
data = f.read()
f.close()
```

```python
data[:100]
```

```python
data = data.split('\n')
```

```python
for line in data : 
    if line : 
        y  = line.split('::')[1][-5:-1]
        if y not in years :         
            years.append(y)
    
```

```python
years.sort()
```

```python
years
```

```python
for key,value in movies.items() : 
    for year in years : 
        value.setdefault(year,[])
```

```python
data = data[:-1]
```

```python
for line in data : 
    movie_name = line.split("::")[1][:-6]
    movie_year = line.split("::")[1][-5:-1]
    geners = line.split('|')
    print(geners)
    geners[0] = ''.join(geners[0]).split("::")[-1]
    print(geners)
    #geners[-1] = geners[-1][:-1]
    for gener in geners : 
        movies[gener][movie_year].append(movie_name)
```
